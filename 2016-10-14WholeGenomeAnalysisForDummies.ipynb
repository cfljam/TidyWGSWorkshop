### Calculate Windowed Nucleotide Diversity in 10kb Windows from Freebayes

- use CHR1 variants called by Freebayes
- Use one of the options in [vcftools](https://vcftools.github.io/man_latest.html)
- run on the PFR job scheduler

````bash
bsub -J CalcPiCHR1 "vcftools \
    --gzvcf ${VCF} \
    --max-missing 1.0 \
    --window-pi 10000 \
    --stdout | gzip -c > CK_13_CHR1.Pi.gz"
````

### Now do the Same for All Chromosomes

- run in parallel on job scheduler
- could do this with gnu parallel

````bash
mkdir PI
````

````bash
VCFDIR=/output/genomic/plant/Actinidia/chinensis/Resequencing/Variants/PS1.1.68.5/13Monkeys/Freebayes
OUTDIR=./PI

for CHR in {1..29}
    do 
    VCFFILE=${VCFDIR}/13Monkeys_CHR${CHR}_PS1_1.68.5_bwaMEM.vcf.gz
    OUTFILE=${OUTDIR}/CHR${CHR}.Pi.gz
    echo "vcftools \
    --gzvcf ${VCFFILE} \
    --max-missing 1.0 \
    --window-pi 10000 \
    --stdout | gzip -c > ${OUTFILE}"
    done | asub
````

### Extract First 0.2 Mbp  of one Chromosome with Vcftools, no missing data

and compress  and index the output

````bash
bsub -J ExtractCHR1 "vcftools \
    --gzvcf ${VCF} \
    --chr CHR1 \
    --to-bp 200000 \
    --recode --recode-INFO-all \
    --max-missing 1.0 \
    --stdout | bgzip -c > CHR1_200kb_PS1_1.68.5_bwaMEM_Freebayes.vcf.gz
    tabix CHR1_200kb_PS1_1.68.5_bwaMEM_Freebayes.vcf.gz"
````

## Access Some Fields Using Shell and Vcflib Commands

### Get the first three multi-allelic records using shell

````bash
zcat ./CHR1_200kb_PS1_1.68.5_bwaMEM_Freebayes.vcf.gz | grep -v "^#" | grep -v "NUMALT=1" | head -n 3
````

### Peek at the sample names using a vcflib command


RTFD @ https://github.com/vcflib/vcflib

````bash
vcfsamplenames ./CHR1_200kb_PS1_1.68.5_bwaMEM_Freebayes.vcf.gz
````

## Filter out Highly Polymorphic Sites

````bash
vcffilter -f 'NUMALT > 3' CHR1_200kb_PS1_1.68.5_bwaMEM_Freebayes.vcf.gz |  \
    vcfkeepgeno - GT | grep -v "^#" | head -n 5
````
